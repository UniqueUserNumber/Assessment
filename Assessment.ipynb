The Point of this notebook below is to 
- Create 50000 random numbers and then do a univariate analysis on these numbers.
- Take 15% of these randomly generate numbers and do a univariate analysis on these numbers.
- Compare these two Univariate analysis

```python
import numpy as np
import pandas as pd
import seaborn as sns
from scipy.stats import ttest_ind
```

```python
# Create the 50000 and the subset datasets
data = np.random.randint(0, 101, size=50000)
df_Big = pd.DataFrame(data, columns=['Numbers'])
df_Small = df_Big.sample(frac=0.15, replace=False,  random_state=1)
df_Small.reset_index(drop=True)
```

```python
df_Big['Numbers'].value_counts().sort_index().plot.line()
```

```python
plot_Big = df_Big['Numbers'].value_counts().sort_index().plot.line()
plot_Big.set_ylim(0,50000/70)
```

```python
sns.distplot(df_Big)
```

```python
plot_Small = df_Small['Numbers'].value_counts().sort_index().plot.line()
```

```python
plot_Small = df_Small['Numbers'].value_counts().sort_index().plot.line()
plot_Small.set_ylim(0,7500/70)
```

```python
sns.distplot(df_Small)
```

```python
#Univariate Analysis on 50,000 Random Numbers
Big_Mean = df_Big.mean()
Big_Median = df_Big.median()
Big_Variance = df_Big.var()
Big_STD = df_Big.std()

print('The Mean for the original dataset is ' + str(round(Big_Mean[0],2)))
print('The Median for the original dataset is ' + str(round(Big_Median[0],2)))
print('The Variance for the original dataset is ' + str(round(Big_Variance[0],2)))
print('The Standard Deviation for the original dataset is ' + str(round(Big_STD[0],2)))
```

```python
#Univariate Analysis on 15% sample of Prior Set
Small_Mean = df_Small.mean()
Small_Median = df_Small.median()
Small_Variance = df_Small.var()
Small_STD = df_Small.std()


print('The Mean for the sample data set is ' + str(round(Small_Mean[0],2)))
print('The Median for the sample data set is ' + str(round(Small_Median[0],2)))
print('The Variance for the sample data set is ' + str(round(Small_Variance[0],2)))
print('The Standard Deviation for the sample data set is ' + str(round(Small_STD[0],2)))
```

```python
#Comparison between the two datasets
#Correlation = df_Big.corrwith(df_Small, axis = 0) 
#Correlation[0]

t_test = ttest_ind(df_Big, df_Small)
print('The t-value of the original data set and the sample data set are ' + str(round(t_test[0][0],4)))
print('The p-value of the original data set and the sample data set are ' + str(round(t_test[1][0],4)))
```
